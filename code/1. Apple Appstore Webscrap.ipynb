# Apple Appstore webscrap

```python
from app_store_scraper import AppStore
import pandas as pd
```

```python
shoppee = AppStore(country = 'sg', app_name = 'shoppee')
shoppee.review(how_many = 20)
```

```python
lazada = AppStore(country = 'sg', app_name = 'lazada')
lazada.review(how_many = 20)
```

```python
df_shoppee = pd.DataFrame(shoppee.reviews)
df_lazada = pd.DataFrame(lazada.reviews)
df_shoppee['app_Id'] = 'shoppee'
df_lazada['app_Id'] = 'lazada'
df = df_shoppee.append(df_lazada, ignore_index = True)
df.head()
```

```python
app_packages = ['shoppee', 
                'lazada', 
                'carousell',
               'ebay',
               'amazon',
               'asos',
               'wish',
               'aliexpress']

app_id = [959840394, 
          785385147, 
          548607187, 
          282614216, 
          297606951, 
          457876088, 
          530621395, 
          436672029]

for i in range(len(app_packages)):
    name = app_packages[i]
    app = app_packages
    app[i] = AppStore(country = 'sg', app_name = name, app_id = app_id[i], log_interval = 5)
    app[i].review(how_many = 10000)
    app_packages[i] =  pd.DataFrame(app_packages[i].reviews)
    app_packages[i]['appid'] = name
    
df = app_packages[0]
for j in range(1, len(app_packages)):
        df = df.append(app_packages[j])

df.reset_index(inplace = True, drop = True)
```
