```python
import json
import pandas as pd
from tqdm import tqdm

import seaborn as sns
import matplotlib.pyplot as plt

from pygments import highlight
from pygments.lexers import JsonLexer
from pygments.formatters import TerminalFormatter
import matplotlib.font_manager as fm

from google_play_scraper import Sort, reviews, app #https://github.com/JoMingyu/google-play-scraper

%matplotlib inline
%config InlineBackend.figure_format='retina'

sns.set(style='whitegrid', palette='muted', font_scale=1.2)
```

```python
app_packages = [
    'com.shopee.sg',
    'com.thecarousell.Carousell',
    'com.lazada.android',
    'com.alibaba.aliexpresshd',
    'com.ebay.mobile',
    'com.amazon.mShop.android.shopping',
    'com.asos.app',
    'com.contextlogic.wish'
]
```

```python
app_infos = []

for ap in tqdm(app_packages):
    info = app(ap, lang='en', country='sg')
    del info['comments']
    app_infos.append(info)
```

```python
def print_json(json_object):
    json_str = json.dumps(
    json_object,
    indent=2,
    sort_keys=True,
    default=str
    )
    print(highlight(json_str, JsonLexer(), TerminalFormatter()))
```

```python
print_json(app_infos[2])
```

```python
def format_title(title):
    sep_index = title.find(':') if title.find(':') != -1 else title.find('-')
    if sep_index != -1:
        title = title[:sep_index]
    return title[:10]

fig, axs = plt.subplots(2, len(app_infos) // 2, figsize=(14, 5))

for i, ax in enumerate(axs.flat):
    ai = app_infos[i]
    img = plt.imread(ai['icon'])
    ax.imshow(img)
    ax.set_title(format_title(ai['title']))
    ax.axis('off')
```

```python
app_infos_df = pd.DataFrame(app_infos)
app_infos_df.to_csv('shoppingapps_info.csv', index=None, header=True)
```

```python
app_reviews = []

for ap in tqdm(app_packages):
    for sort_order in [Sort.MOST_RELEVANT, Sort.NEWEST]:
        rvs, _ = reviews(ap,lang='en',
                         country='sg',
                         sort=sort_order,
                         count= 10000
                        )
    for r in rvs:
        r['sortOrder'] = 'most_relevant' if sort_order == Sort.MOST_RELEVANT else 'newest'
        r['appId'] = ap
    app_reviews.extend(rvs)
```

```python
len(app_reviews)
```

```python
app_reviews_df = pd.DataFrame(app_reviews)
app_reviews_df.to_csv('google_apps_reviews.csv', index=None, header=True)
```
